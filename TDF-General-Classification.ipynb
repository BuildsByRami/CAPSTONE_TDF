### Test on 2021 data, train on the 10 years prior

```python
train = df[df.Year != 2021]
test = df[df.Year == 2021]
```

```python
test
```

```python
X_train = train.drop(['winner'], axis=1)
y_train = train.winner
```

```python
X_test = test.drop(['winner'], axis=1)
y_test = test.winner
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
print(y_train.shape)
print(y_test.shape)
```

### Importing our basic models
#### More advance models should come later

```python
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
```

### Lets create a validation set to properly tune our models and avoid over fitting

```python
from sklearn.model_selection import train_test_split
```

```python
X_train_tt, X_val_tt, y_train_tt, y_val_tt = train_test_split(X_train, y_train, test_size=0.2, stratify = y_train)
```

# Logistic Regression

```python
# Optimizing our Hyperparameters to get the best results
train_scores = []
validation_scores = []

C_range = np.array([.00000001,.0000001,.000001,.00001,.0001,.001,0.1,\
                1,10,100,1000,10000,100000,1000000,10000000,100000000,1000000000])

for c in C_range:
    my_logreg = LogisticRegression(C = c,random_state=1, max_iter=1000)
    my_logreg.fit(X_train_tt, y_train_tt)
    
    # train on traning set
    train_scores.append(my_logreg.score(X_train_tt,y_train_tt))
    validation_scores.append(my_logreg.score(X_val_tt,y_val_tt))
```

```python
# Plotting the training and validation scores
plt.figure(figsize=(15,5))

plt.plot(C_range, train_scores, label="Train Score",marker='.')
plt.plot(C_range, validation_scores, label="Val Score",marker='.')

plt.xscale('log')
plt.title('Training score vs Validation score over a range of C values')
plt.xlabel('C')
plt.ylabel('Accuracy')
plt.axvline(1, linestyle = '--')
plt.legend()
plt.show()
```

```python
# Instatiate with optimal hyperparameters
log_reg = LogisticRegression(C=1,random_state=1, max_iter=1000)

# Fitting the model to the complete train/validate set
log_reg.fit(X_train, y_train)
print(log_reg.score(X_train,y_train))
print(log_reg.score(X_test,y_test))
```

### Confusion Matrix for the LR model

```python
from sklearn.metrics import plot_confusion_matrix, confusion_matrix
```

```python
# Get class predictions
y_pred = my_logreg.predict(X_test)

# Generate confusion matrix
cf_matrix = confusion_matrix(y_test, y_pred)

# label rows and columns
cf_df = pd.DataFrame(
    cf_matrix, 
    columns=["Predicted 0", "Predicted 1"],
    index=["True 0", "True 1"]
)

display(cf_df)
```

# Decision Tree

```python
# A list of the maximum depths to try out
depths = range(1, int(np.sqrt(X_train.shape[0])))
```

```python
# Empty lists to append to
train_DT = []

# Loop through the depths
for max_depth in depths:
    
    # Instantiate the model 
    DT = DecisionTreeClassifier(max_depth=max_depth, random_state=11)
    
    # Fit the model 
    DT.fit(X_train, y_train)
    
    # Score the model 
    train_DT.append(DT.score(X_train, y_train))
```

```python
# Plotting the accuracy scores
plt.figure(figsize=(15,5))
plt.plot(depths, train_DT, c='red', label='train')
plt.xlabel("Max Depth")
plt.ylabel("Accuracy Score")
plt.title('Testing Tree Depth for Decision Trees Using MinMax Scaler')
plt.axvline(9, linestyle = '--')
plt.legend()
plt.show()
```

```python
# Instantiating the model with optimized hyper parameters
DT = DecisionTreeClassifier(max_depth=9, random_state=42)

# Fit the model 
DT.fit(X_train, y_train)

# Store accuracy scores on full training set and test set
DT_train = DT.score(X_train, y_train)
DT_test = DT.score(X_test, y_test)

# Score the model 
print(f"Train Accuracy: {DT_train}")
print(f"Test Accuracy: {DT_test}")
```

```python
def PlotBoundaries(model, X, Y, dot_size=20, figsize=(10,7)) :
    '''
    Helper function that plots the decision boundaries of a model and data (X,Y)
    code modified from: https://scikit-learn.org/stable/auto_examples/neighbors/plot_classification.html
    '''
    
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1,X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01), np.arange(y_min, y_max, 0.01))

    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.figure(figsize=figsize)
    plt.contourf(xx, yy, Z, alpha=0.4)

    #Plot
    plt.scatter(X[:, 0], X[:, 1], c=Y, s=dot_size, edgecolor='k')
    plt.show()
```

## Lets only look at winners vs rest and use SMOTE to fix our imbalanced data issue

```python
#SMOTE
from imblearn.over_sampling import SMOTE
```

```python
X_train_sm, y_train_sm = SMOTE(random_state=1).fit_resample(X_train_tt, y_train_tt)
```

```python
print('Original')
display(pd.Series(y_train).value_counts().sort_index())
print('Resampled')
display(pd.Series(y_train_sm).value_counts().sort_index())
```

### Logistic Regression Model with class imbalance addressed

```python
# Optimizing our Hyperparameters to get the best results
train_scores = []
validation_scores = []

C_range = np.array([.00000001,.0000001,.000001,.00001,.0001,.001,0.1,\
                1,10,100,1000,10000,100000,1000000,10000000,100000000,1000000000])

for c in C_range:
    my_logreg = LogisticRegression(C = c,random_state=1, max_iter=1000)
    my_logreg.fit(X_train_sm, y_train_sm)
    
    # train on traning set
    train_scores.append(my_logreg.score(X_train_sm,y_train_sm))
    validation_scores.append(my_logreg.score(X_val_tt,y_val_tt))
```

```python
# Plotting the training and validation scores
plt.figure(figsize=(15,5))

plt.plot(C_range, train_scores, label="Train Score",marker='.')
plt.plot(C_range, validation_scores, label="Val Score",marker='.')

plt.xscale('log')
plt.title('Training score vs Validation score over a range of C values')
plt.xlabel('C')
plt.ylabel('Accuracy')
plt.axvline(0.001, linestyle = '--')
plt.legend()
plt.show()
```

### Scores on final model

```python
# Instatiate with optimal hyperparameters
log_reg = LogisticRegression(C=0.001,random_state=1, max_iter=1000)

# Fitting the model to the complete train/validate set
log_reg.fit(X_train_sm, y_train_sm)
print(log_reg.score(X_train_sm,y_train_sm))
print(log_reg.score(X_test,y_test))
```
