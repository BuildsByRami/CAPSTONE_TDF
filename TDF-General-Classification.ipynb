```python
df = df[df.Rank < 101]
```

```python
df = df.drop(['Prev_rank','Team_name','Points','Time'], axis=1)
```

```python
df.head(2)
```

```python
df = pd.get_dummies(df, columns=['Rider_name'], prefix='Rider')

#Glimpse at the modified data
df.head(2)
```

```python
df.loc[df.Rank == 1, 'winner'] = 1

df.head()
```

```python
df = df.fillna(0)
```

```python
df.winner = df.winner.astype('int8')
```

```python
df = df.drop('Rank',axis = 1)
df.head(2)
```
